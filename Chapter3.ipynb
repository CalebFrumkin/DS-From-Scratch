```python
years = [10 * x + 1950 for x in range(7)]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

plt.title('Nominal GDP')
plt.ylabel("Billions of $")
plt.show()
```

```python
from collections import Counter
import random
```

```python
random.seed(5)
grades = [random.randrange(101) for _ in range(100)]

histogram = Counter(min(grade // 10*10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()],
        histogram.values(),
        10,
        edgecolor=(0,0,0))

plt.axis([-5, 105, 0, max(histogram.values()) + 5])

plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of Students')
plt.title('Distribution of Exam 1 Grades')
```

```python
variance = [2**x for x in range(9)]
biasSquared = variance.copy()
biasSquared.reverse()

totalError = [x + y for x, y in zip(variance, biasSquared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, biasSquared, 'r-.', label='bias^2')
plt.plot(xs, totalError, 'b:', label='total error')

plt.legend(loc=9)
plt.xticks([])
```

```python
friends = [random.randrange(60,101) for _ in range(9)]
minutes = [random.randrange(100, 250) for _ in range(9)]
labels = ['a','b','c','d','e','f','g','h','i']

plt.scatter(friends, minutes)

for label, friendCount, minCount in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friendCount, minCount),
                 xytext=(5,-5),
                 textcoords='offset points')
    
    plt.title('Daily Mins vs. # of Friends')
    plt.xlabel('# of Friends')
    plt.ylabel('daily mins on site')
```
